```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0, 10, 0.1)
sin = np.sin(x)
cos = np.cos(x)
```

```python

plt.plot(x, sin, label='sin')
plt.plot(x, cos, label='cos')
plt.title('sin and cos curve')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()
plt.show()
```

```python
# First create some toy data:
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)

# Create just a figure and only one subplot
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('Simple plot')

# Create two subplots and unpack the output array immediately
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
ax1.plot(x, y)
ax1.set_title('Sharing Y axis')
ax2.scatter(x, y)
```

```python
#  13
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 3))

# ax1 
ax1.bar([1,2,3], [3,4,5])

# ax2 
ax2.barh([0.5, 1, 2.5], [0,1,2])

# ax3 
ax3.plot(x, sin, label='sin')
ax3.set_title('sin')
ax3.set_xlabel('Time')
ax3.set_ylabel('Amplitude')

plt.legend()
plt.tight_layout()
plt.show()
```

```python
# (12)
fig, ax = plt.subplots(2, 2, figsize=(12, 6))

# 
ax[0][0].bar([1,2,3], [3,4,5], color='g')
ax[0][1].barh([0.5, 1, 2.5], [0,1,2])
ax[1][0].plot(x, sin, label='sin', linestyle='--', linewidth='5')
ax[1][1].scatter(x, cos, label='cos', marker='.')

plt.legend(loc='center left')
plt.tight_layout()
plt.show()
```
